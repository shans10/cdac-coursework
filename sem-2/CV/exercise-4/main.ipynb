```python
import skimage as sk
import matplotlib.pyplot as plt
image = sk.io.imread("../image.png", as_gray=True)
plt.title("Original Image")
plt.imshow(image)
```

#### 1. Shear

```python
shear = sk.transform.warp(image,sk.transform.AffineTransform(shear=0.5))
plt.title("Sheared Image")
plt.imshow(shear)
```

#### 2. Scale

```python
scaled = sk.transform.rescale(image, 0.5)
plt.title("Scaled Image")
plt.imshow(scaled)
plt.show()
```

#### 3. Transition

```python
transition = sk.transform.warp(image, sk.transform.AffineTransform(translation=(100, 50)))
plt.title("Translated Image")
plt.imshow(transition)
plt.show()
```

#### 4. Rotation

```python
rotated = sk.transform.rotate(image, 65)
plt.title("Rotated Image")
plt.imshow(rotated)
```

#### 5. Affine

```python
affined = sk.transform.warp(image, sk.transform.AffineTransform(translation=(1,2), shear=0.2))

plt.figure(figsize=(5, 5))
plt.title("Affine Transformed Image")
plt.imshow(affined)
plt.show()
```
